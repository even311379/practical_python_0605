```python
d1 = Digraph(engine="neato")
d1.attr(size='5!,4')
d1.attr('node',shape = 'Msquare')
d1.node('p','電腦',pos='0,0!')
d1.node('h','人類',pos='3,0.5!')
d1.attr('edge', fontsize='10')
d1.edge('h','p','程式語言',constraint='false')
d1.edge('p','p','理解')
d1
```

##### 高階 VS 低階程式語言

```python
d = Digraph(engine="neato")
d.attr(size='5!,4')
d.attr('node',shape = 'box')
d.node('l','低階',pos='0,0!')
d.node('h','高階',pos='4,0!')
d.edge('l','h',dir='both')
d.attr('node',shape = 'plaintext')
d.node('le','電腦容易理解',pos='0,0.5!')
d.node('he','人類容易理解',pos='4,0.5!')
d.edge('le','l',dir='none')
d.edge('he','h',dir='none')

d.attr('node',shape = 'underline')
d.node('c0','machine code',pos='0,-1!')
d.node('c1','assembly',pos='1.2,-1!')
d.node('c2','c/c++',pos='2.2,-1!')
d.node('c3','Java',pos='3,-1!')
d.node('c4','Python',pos='5,-1!')
d
```
